```python
import matplotlib.pyplot as plt
import numpy as np

# Explanation of plt.subplot()
# The plt.subplot() function takes three arguments: nrows, ncols, and index.
# - nrows: The number of rows in the grid of subplots.
# - ncols: The number of columns in the grid of subplots.
# - index: The index of the current subplot (starting from 1 in the top-left).

# It adds an Axes object to the current figure, which acts as a plotting area for a single subplot.
# If you call it multiple times with different indices, it will create multiple subplots.

# Example:
# Create a figure with a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# First subplot (top-left)
plt.subplot(2, 2, 1)
plt.plot(np.random.rand(10), color='red') #10 random numbers generated between 0 and 1
plt.title('Subplot 1')

# Second subplot (top-right)
plt.subplot(2, 2, 2)
plt.plot(np.sin(np.linspace(0, 2*np.pi, 100)), color='blue')
plt.title('Subplot 2')

# Third subplot (bottom-left)
plt.subplot(2, 2, 3)
plt.hist(np.random.randn(100), bins=5, color='green')
plt.title('Subplot 3')

# Fourth subplot (bottom-right)
plt.subplot(2, 2, 4)
plt.scatter(np.random.rand(20), np.random.rand(20), color='purple')
plt.title('Subplot 4')

plt.tight_layout() # Adjusts subplot parameters for a tight layout
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='scatter')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
#Categorical Data plots
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x= 'tip',y = 'day',data = df,palette = 'rainbow')
```

```python

# boxplot and violinplot
# boxplots and violinplots are used to shown the distribution of categorical data.
# A box plot (or box-and-whisker plot) shows the distribution of quantitative data
# in a way that facilitates comparisons between variables or across levels of a categorical variable.
# The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution,
# except for points that are determined to be outliers using a method that is a function of the inter-quartile range.
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python

sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```
