```python
#A series is a one dimensional labelled array capable of holding any data type .The axis labels are collectively called the index
# There are multiple ways to create a pandas series.
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d={1:10,2:20,3:30}
```

```python
pd.Series(d)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
#Creating a Dataframe
data={
    'Name':['Jai','Princi','Gaurav','Anuj'],
    'Age':[27,24,22,32],
    'Address':['Delhi','Kanpur','Allahabad','Kannauj'],
    'Qualification':['Msc','MA','MCA','Phd']
}
df=pd.DataFrame(data)
df
```

```python
data_list=[['Jai',27],['Princi',24],['Gaurav',22],['Anuj',23]]
df2=pd.DataFrame(data_list)
columns=['Name','Age']
df2 = pd.DataFrame(data_list,columns=columns)
df2
```

```python
#Selection and Indexing of Columns
df2[['Age','Name']]
```

```python
#Creating a new Column
df2['Address']=['Delhi','Kanpur','Allahabad','Kannauj']
df2
```

```python
#Removing Column
df2.drop('Address',axis=1,inplace=True)
df2
```

```python
#Selecting Rows
df2.loc[0]
df2.iloc[2]
df2
```

```python
#Selecting Subsets of Rows and Columns
df2.loc[[0,1]][["Name"]]
```

```python
#Conditional Selection
#I want only those people whose age is above 25
df2[df2["Age"]>25]
```

```python
#I want people whose age is above 25 and their name is Suhani
df2[(df2["Age"]<25) & (df2["Name"] == 'Princi')]
```
